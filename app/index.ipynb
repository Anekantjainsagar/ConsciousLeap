```python
from bs4 import BeautifulSoup 
import requests 
```

```python
URL = "https://ngodarpan.gov.in/index.php/home/statewise_ngo/18730/9/4" 
r = requests.get(URL) 
print(r.content) 
```

```python
soup = BeautifulSoup(r.content) # If this line causes an error, run 'pip install html5lib' or install html5lib 
print(soup.prettify()) 
```

```python
table = soup.findAll('tr')[1:]
```

```python
for i in range(len(table)):
    name = table[i].find_all('td')[1].text.strip()
    print(f"Name: {name}")
    print("")
```
